# Question -2. Create 1 row and 3 column for subplot the binary, grey and color image.

# Question -3. Add text and different shapes in image and display the original as well as output image.

# Question -7. Reflect the input image both vertically and horizontally. 

```python
# Load the image in BGR format
bgr_image = cv2.imread('teddy.jpg')

# Reflect the image horizontally
reflected_image1 = cv2.flip(bgr_image, 1)

# Reflect the image vertically 
reflected_image2 = cv2.flip(bgr_image, 0)

# Reflect the image both vertically and horizontally
reflected_image3 = cv2.flip(bgr_image, -1)


# Create a subplot with 1 row and 4 columns
plt.figure(figsize=(12, 4))
plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(cv2.cvtColor(reflected_image1, cv2.COLOR_BGR2RGB))
plt.title('Horizonal Reflection')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(cv2.cvtColor(reflected_image2, cv2.COLOR_BGR2RGB))
plt.title('Vertical Reflection')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(reflected_image3, cv2.COLOR_BGR2RGB))
plt.title('Horizontal and Vertical Reflection')
plt.axis('off')

plt.show()
```

# Question -8. Rotate the image with the following instruction 

## a) Rotate it from the centre of the image, theta is 90, scale down by 0.5 
## b) Rotate it from the centre of the image, theta is 90, upscale by 2.5 
## c) Observe the difference between upscale and downscale

```python
# Load the image in BGR format
bgr_image = cv2.imread('flower.jpg')

# Define rotation angles and scale factors
angle1 = 90
angle2 = 90
scale_down = 0.5
scale_up = 2.5

# Get the center of the image
height, width = bgr_image.shape[:2]
center = (width // 2, height // 2)

# Rotate the image from the center with different angles and scale factors
rotated_down = cv2.warpAffine(bgr_image,cv2.getRotationMatrix2D
                              (center, angle1, scale_down), (width, height))
rotated_up = cv2.warpAffine(bgr_image, cv2.getRotationMatrix2D
                            (center, angle2, scale_up), (width, height))

# Create a subplot with 1 row and 3 columns
plt.figure(figsize=(15, 10))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(rotated_down, cv2.COLOR_BGR2RGB))
plt.title('Rotated 90°, Scale Down 0.5')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(rotated_up, cv2.COLOR_BGR2RGB))
plt.title('Rotated 90°, Scale Up 2.5')
plt.axis('off')

plt.show()
```

# Question -9. Find the distance between two pixels in the image using,
## a) Distance Transform
## b) Euclidean Distance
## c) City block Distance
## d) Chess board Distance

```python
# Load a color image
image = cv2.imread('lake.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
```

```python
# Define the coordinates of two points
x1, y1 = 100, 100
x2, y2 = 300, 500

# Calculate distance using Distance Transform
distance_transform = cv2.distanceTransform(image, cv2.DIST_L2, 3)
distance_dt = distance_transform[y1, x1] - distance_transform[y2, x2]
print(f"Distance using Distance Transform: {distance_dt}")
```

```python
# Calculate Euclidean Distance
point1 = np.array([x1, y1])
point2 = np.array([x2, y2])
euclidean_distance = np.linalg.norm(point1 - point2)
print(f"Euclidean Distance: {euclidean_distance}")
```

```python
# Calculate City Block Distance
city_block_distance = np.sum(np.abs(point1 - point2))
print(f"City Block Distance: {city_block_distance}")
```

```python
# Calculate Chessboard Distance
chessboard_distance = np.max(np.abs(point1 - point2))
print(f"Chessboard Distance: {chessboard_distance}")
```
